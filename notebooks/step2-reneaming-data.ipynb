```python
def add_category_prefix(filenames, data_dict):
    def process_single_file(filename):
        # Remove o prefixo da categoria, se existir
        if filename.startswith(("stretchedSociety_", "establishedAffluence_")):
            _, filename = filename.split("_", 1)
        
        # Verifica em qual categoria o arquivo se encaixa
        if filename in data_dict["stretchedSociety"]:
            return "stretchedSociety_" + filename
        elif filename in data_dict["establishedAffluence"]:
            return "establishedAffluence_" + filename
        else:
            return filename  # Retorna o nome original se não pertencer a nenhuma categoria

    # Se filenames for uma string (caminho de diretório), converte para lista de arquivos
    if isinstance(filenames, str):
        filenames = os.listdir(filenames)

    # Processa cada arquivo na lista
    return [process_single_file(filename) for filename in filenames]
```

```python
results = add_category_prefix(os.listdir("../EDA/tft_houses/"), data_dict)

# Ou assim, se quiser processar todos os arquivos em um diretório:
# results = add_category_prefix("../EDA/tft_houses/", data_dict)

for original, result in zip(os.listdir("../EDA/tft_houses/"), results):
    print(f"Original: {original}")
    print(f"Resultado: {result}")
    print()
```

```python
def rename_files_with_category(directory, data_dict):
    def get_category_prefix(filename):
        if filename in data_dict["stretchedSociety"]:
            return "stretchedSociety_"
        elif filename in data_dict["establishedAffluence"]:
            return "establishedAffluence_"
        else:
            return ""  # Retorna vazio se não pertencer a nenhuma categoria

    renamed_files = []
    errors = []

    for filename in os.listdir(directory):
        if filename.startswith(("stretchedSociety_", "establishedAffluence_")):
            continue  # Pula arquivos que já têm prefixo

        old_path = os.path.join(directory, filename)
        prefix = get_category_prefix(filename)
        
        if prefix:
            new_filename = prefix + filename
            new_path = os.path.join(directory, new_filename)
            
            try:
                os.rename(old_path, new_path)
                renamed_files.append((filename, new_filename))
            except OSError as e:
                errors.append((filename, str(e)))
        else:
            errors.append((filename, "Categoria não encontrada"))

    return renamed_files, errors
```

```python
directory = "../EDA/tft_houses/"
renamed, errors = rename_files_with_category(directory, data_dict)

print("Arquivos renomeados:")
for old, new in renamed:
    print(f"{old} -> {new}")

print("\nErros:")
for file, error in errors:
    print(f"{file}: {error}")
```
