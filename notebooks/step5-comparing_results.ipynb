```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../src_Prophet_combinations/results/all_combinations_results.csv')
```

```python
df.head()
```

```python
df.groupby(['dataset','model']).max()
```

```python

sns.set_theme(style="ticks")
```

```python
for house in df.dataset.unique():
    casas = df[df['dataset'].isin([house])]
    
    f, ax = plt.subplots(figsize=(7, 6))

    sns.boxplot(
        casas, x="mae", y="dataset", hue="dataset",
        whis=[0, 100], width=.6, palette="vlag"
    )

    # Add in points to show each observation
    sns.stripplot(casas, x="mae", y="dataset", size=4, color=".3")

    # Tweak the visual presentation
    ax.xaxis.grid(True)
    ax.set(ylabel="")
    sns.despine(trim=True, left=True)
```
